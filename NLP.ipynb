```python
import openai
import spacy
import PyPDF2
import re

# Set up your OpenAI API credentials
openai.api_key = 'gggg'  #here I removed my key, please use your own openAI API key.

def generate_mcq_question(paragraph):
    prompt = f"Generate a multiple-choice question based on the following paragraph:\n{paragraph}\nOptions:"
    
    response = openai.Completion.create(
        engine='text-davinci-003',
        prompt=prompt,
        temperature=0.5,
        max_tokens=128,
        n=4,
        stop=None
    )
    
    question = response.choices[0]['text']
    
    return question

def generate_mcq_questions_with_options(text):
    paragraphs = text.split('\n\n')
    mcq_questions = []
    mcq_options = []
    
    for paragraph in paragraphs:
        question = generate_mcq_question(paragraph)
        options_prompt = f"{question}\nOptions:"
        
        response = openai.Completion.create(
            engine='text-davinci-003',
            prompt=options_prompt,
            temperature=0.5,
            max_tokens=128,
            n=4,
            stop=None
        )
        
        options = [choice['text'].strip() for choice in response.choices]
        
        mcq_questions.append(question)
        mcq_options.append(options)
    
    return mcq_questions, mcq_options

texts=''
chapters=['chapter-2.pdf','chapter-3.pdf','chapter-4.pdf']
for chapter in chapters:
    with open('chapter-2.pdf', 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)

        start_page = 1  
        end_page = 15 

        for page_number in range(start_page, end_page + 1):
            page = pdf_reader.pages[page_number - 1]  # Adjust index for zero-based indexing

            # Extract the text content of the page
            text = page.extract_text()

            # Split the text into paragraphs
            paragraphs = text.split('\n\n')  # Adjust the delimiter as per your PDF's paragraph structure


            # Iterate through the desired paragraphs and print them
            text=''
            for paragraph in paragraphs:
                pattern = r"^(?:Fig|chap).*"

                paragraph = re.sub(pattern, "", paragraph, flags=re.MULTILINE)
                paragraph = paragraph.replace('\n', ' ')
                paragraph = re.sub(r"\s{2,}", " ", paragraph)
                text += paragraph+'\n'

            texts=texts+text+'\n'
            


# Generate MCQ questions with options for each paragraph
generated_mcq_questions, generated_mcq_options = generate_mcq_questions_with_options(texts)

# Print the generated MCQ questions with options
for i, question in enumerate(generated_mcq_questions, start=1):
    print(f"Q{i}: {question}")
    print("Options:")
    for j, option in enumerate(generated_mcq_options[i-1], start=1):
        print(f"{j}. {option}")
    print()
```
